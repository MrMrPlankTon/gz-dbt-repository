```python
df_2020 = pd.read_csv('/content/full (5).csv')
df_2020
```

```python
#supprimer les Nan dans surface réelle - enlève aussi les doublons
df_2020 = df_2020.dropna(subset=['type_local', 'valeur_fonciere', 'surface_reelle_bati', 'nombre_pieces_principales'])
df_2020
```

```python
# group by dept puis type_local, SUM surface_relle_bati, SUM valeur_fonciere, prix_m2
df_2020['num_departement'] = (df_2020['code_commune']/1000).astype(int64)
df_dept_2020 = df_2020.groupby(['num_departement', 'type_local']).agg(
    SUM_surface_reelle_bati = ('surface_reelle_bati', 'sum'),
    SUM_valeur_fonciere = ('valeur_fonciere', 'sum')
).reset_index()
df_dept_2020['prix_m2'] = (df_dept_2020['SUM_valeur_fonciere'] / df_dept_2020['SUM_surface_reelle_bati']).astype(int64)
df_dept_2020
```

```python
df_2024 = pd.read_csv('/content/full (1).csv')
df_2024
```

```python
df_2025 = pd.read_csv('/content/full.csv')
df_2025
```

```python
df_total = pd.concat([df_2020, df_2021, df_2022, df_2023, df_2024, df_2025], ignore_index=True)
df_total
```

```python
df_total['date_mutation'] = pd.to_datetime(df_total['date_mutation'], format='%Y-%m-%d').dt.date
df_total['valeur_fonciere'] = df_total['valeur_fonciere'].fillna(0).astype('int64')
df_total['code_postal'] = df_total['code_postal'].fillna(0).astype('int64')
df_total['num_departement'] = (df_total['code_postal']/1000).astype('int64')
df_total['surface_reelle_bati'] = df_total['surface_reelle_bati'].fillna(0).astype('int64')
df_total['nombre_pieces_principales'] = df_total['nombre_pieces_principales'].fillna(0).astype('int64')
df_total['surface_terrain'] = df_total['surface_terrain'].fillna(0).astype('int64')
```

```python
#supprimer les Nan dans surface réelle - enlève aussi les doublons
df_total = df_total.dropna(subset=['type_local', 'valeur_fonciere', 'surface_reelle_bati', 'nombre_pieces_principales'])
df_total
```

```python
# garder les colonnes (id_mutation, date_mutation, nature_mutation, valeur_fonciere, adresse_numero, adresse_nom_voie,
# code_postal, code-commune, nom_commune, code-departement, nombre_lots, surface_reelle_bati,
# nombre_pieces_principales, surface_terrain, longitude (WGS-84), latitude(WGS-84))
df_filre_doublons = df_total.drop_duplicates(subset=['id_mutation', 'valeur_fonciere'])
df_grouped = df_filre_doublons.groupby('id_mutation').agg(
    date_mutation=('date_mutation', 'first'),
    nature_mutation=('nature_mutation', 'first'),
    valeur_fonciere=('valeur_fonciere', 'first'),
    code_postal=('code_postal', 'first'),
    code_commune=('code_commune', 'first'),
    commune=('nom_commune', 'first'),
    departement=('code_departement', 'first'),
    nb_lots=('nombre_lots', 'first'),
    surface_batie=('surface_reelle_bati', 'first'),
    nb_pieces=('nombre_pieces_principales', 'first'),
    surface_terrain=('surface_terrain', 'sum'),
    longitude=('longitude', 'first'),
    latitude=('latitude', 'first')
).reset_index()
df_grouped
```
